```python
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head(10)
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=20)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.barplot(x='sex',y='tip',data=tips,estimator=np.sum)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='smoker',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='box',hue='sex',dodge=True)
```
